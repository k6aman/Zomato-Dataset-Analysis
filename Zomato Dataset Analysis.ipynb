```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("zomato data set.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

Dropping Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

Cleaning Rate Column

```python
df['rate'].unique()
```

Removing "NEW" , "-" and "/5" from Rate Column

```python
def handle_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handle_rate)
df['rate'].head()
```

Filling Null Values in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

Dropping Null Values

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Listed in(city) and location, both are same, lets keep only one.

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

Removing "," from Cost2Plates Column

```python
def handle_comma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handle_comma)
df['Cost2plates'].unique()
```

```python
df.head()
```

Cleaning rest_type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_less_than_1000 = rest_types[rest_types<1000]
rest_types_less_than_1000
```

Making rest_type less than 1000 in frequency as 'others'

```python
def handle_rest_type(value):
    if(value in rest_types_less_than_1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False)
location_less_than_300 = location[location<300]

def handle_location(value):
    if(value in location_less_than_300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines_less_than_100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_less_than_100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
